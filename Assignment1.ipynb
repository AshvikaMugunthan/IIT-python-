```python
name=str(input("Enter Your Name:"))
marks=int(input("Enter Your Marks"))
if marks>100:
    print("invalid input")
elif marks<0:
    print("invalid marks")
elif marks>=80:
    print("Name:",name,"",
          "Marks:",marks,"",
          "Gradre:","A")
elif marks>=60:
    print("Name:",name,"",
          "Marks:",marks,"",
          "Gradre:","B")
elif marks>=40:
    print("Name:",name,"",
          "Marks:",marks,"",
          "Gradre:","C")
else:
    print("Fail")
```
